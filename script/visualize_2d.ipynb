```python
import cv2
import glob
img_2d_path=glob.glob('../results/2D/v1_pose0/*.png')
print(len(img_2d_path))
import matplotlib.pyplot as plt
import numpy as np

def cvt(img):
    return cv2.cvtColor(img,cv2.COLOR_BGR2RGB)


def plot_figures(figures, nrows = 1, ncols=1):
    """Plot a dictionary of figures.

    Parameters
    ----------
    figures : <title, figure> dictionary
    ncols : number of columns of subplots wanted in the display
    nrows : number of rows of subplots wanted in the figure
    """
    
    fig, axeslist = plt.subplots(ncols=ncols, nrows=nrows,figsize=(30,30))
    figures_name = sorted(figures,key = lambda x:int(x[2:]))
    pairs = zip(range(len(figures_name)), figures_name)
    
    for ind,title in pairs:
        #print(ind,title)
        axeslist.ravel()[ind].imshow(figures[title], cmap=plt.jet())
        axeslist.ravel()[ind].set_title(title)
        axeslist.ravel()[ind].set_axis_off()
    plt.tight_layout() # optional

def plot_figures_es(figures, nrows = 1, ncols=1):
    plt.figure(figsize=(20,20))
    for i in range(1,len(figures)+1):
        digit = 10**(len(str(i))-1)
        bas = nrows*100+ncols*10
        bas = bas*digit+i
        #print (bas)
        plt.subplot(bas)
        plt.imshow(figures['im%d'%(i-1)])
    

# generation of a dictionary of (title, images)
start = 200
end = 3000
inteval = 100
num = int((end-start)/inteval)
name_format = "../results/2D/v1_pose0/global_map_pose_e_%d.png"

figures = {'im'+str(i): cv2.imread(name_format%i) for i in range(start,end,inteval)}

print(num)
row = 7 if num > 7 else 1
col = int(num/row)
# plot of the images in a figure, with 5 rows and 4 columns
plot_figures(figures,row,col)

plt.show()
'''
index = 0
print(img_2d_path[index])
img = cv2.imread(img_2d_path[index])
plt.imshow(cvt(img))
'''
```
